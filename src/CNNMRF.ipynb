```python
content_dir = 'dataset/photos/'
style_dir = 'dataset/paintings/'
res_base_dir = 'CNNMRF'

if not os.path.exists(res_base_dir):
    os.mkdir(res_base_dir)
    
for style_img in os.listdir(style_dir):
    num = style_img[:-4]
    
    if num in os.listdir(res_base_dir):
        continue
        
    content_images = [x for x in os.listdir(content_dir) if num in x]
    
    for content_img in content_images:
        res_dir = os.path.join(res_base_dir, content_img[:-4])
        transfer_style(content_dir + content_img, style_dir + style_img, res_dir)
```
