```python
content_dir = "dataset/photos"
fake_style_dir = "dataset/fake_real_paintings"
style_dir = "dataset/paintings"
results_dir = "automated-deep-photo-style-transfer/results/"
```

```python
for fake_style_img in os.listdir(fake_style_dir):
    img_name = fake_style_img[:fake_style_img.index('.')]
    
    content_images = [x for x in os.listdir(content_dir) if img_name in x]
    n_content_imgages = len(content_images)
    
    for content_img in content_images:
        img_number = content_img[:content_img.index('.')]
        
        print("Transferring image number", img_number)
        
        # Create result directory
        result_dir = results_dir + "result_" + img_number
        
        if not os.path.exists(result_dir):
            os.mkdir(result_dir)
            
        if 'iterations' in os.listdir(result_dir):
            continue
            
        # Transfer the style
        transfer_style(os.path.join(content_dir, content_img), os.path.join(fake_style_dir, fake_style_img), result_dir)
        
        # Copy content and style images into result folder
        content_name = img_number + '.jpg'
        style_name = img_number + '_style.jpg'
        fake_style_name = img_number + '_fake_style' + fake_style_img[fake_style_img.index('.'):]
        
        shutil.copy2(os.path.join(content_dir, content_img), os.path.join(result_dir, content_name))
        shutil.copy2(os.path.join(style_dir, img_name + '.jpg'), os.path.join(result_dir, style_name))
        shutil.copy2(os.path.join(fake_style_dir, fake_style_img), os.path.join(result_dir, fake_style_name))

        # Show content, style, and result images
        result_img = os.path.join(result_dir, "final_transfer_image.png")

        display(os.path.join(result_dir, content_img))
        display(os.path.join(result_dir, fake_style_name))
        display(result_img)
```
