```python
import os
from PIL import Image
import numpy as np
import warnings

warnings.filterwarnings("ignore")
os.chdir("/content/drive/MyDrive/_Van_Gogh/Doodle")
```

```python
def transfer_style(n_colors, content_img, content_mask, style_img, style_mask, output_dir):
  %%shell
  !python stylize.py \
       --mask_n_colors=$n_colors \
       --content_img=$content_img \
       --target_mask=$content_mask \
       --style_img=$style_img \
       --style_mask=$style_mask \
       --hard_width=256 \
       --output_dir=$output_dir
```

```python
base_dir = '/content/drive/MyDrive/_Van_Gogh/results'

for f in os.listdir(base_dir):
  num = f[f.index('_')+1:]

  if num != '00474':
    continue
  
  output_dir = os.path.join(base_dir, f, 'Doodle')
  if not os.path.exists(output_dir):
    os.mkdir(output_dir)
  else: # if directory is present, it means that the result has already been computed, so we avoid to recompute it
    continue

  content_img= os.path.join(base_dir, f, num + '.jpg')
  content_mask= os.path.join(base_dir, f, num + '_seg.png')
  style_img= os.path.join(base_dir, f, num + '_style.jpg')
  style_mask= os.path.join(base_dir, f, num + '_style_seg_raw.png')

  # count the number of colors in the mask
  mask = Image.open(content_mask)
  n_colors=len(np.unique(np.sum(np.array(mask), axis=2)))

  print("COMPUTING IMAGE " + num)
  
  transfer_style(n_colors, content_img, content_mask, style_img, style_mask, output_dir)
```

```python
import matplotlib.pyplot as plt

def display(img, title=""):
  img = plt.imread(img)
  plt.imshow(img)
  plt.title(title)
  plt.show()

for f in os.listdir(base_dir):
  num = f[f.index('_')+1:]

  content_img = os.path.join(base_dir, f, num + '.jpg')
  style_img = os.path.join(base_dir, f, num + '_style.jpg')
  res_img = os.path.join(base_dir, f, 'Doodle', 'result_final.png')

  # display all images
  print("Image no. " + num)
  display(content_img, "Content")
  display(style_img, "Style")
  display(res_img, "Result")
```
